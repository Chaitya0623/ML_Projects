Importing hte Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
sonar_data = pd.read_csv('/content/drive/MyDrive/ML_Projects/sonarData.csv', header=None)
sonar_data
```

```python
sonar_data.describe() # statistical measures of the data
```

```python
sonar_data[60].value_counts()
```

M => Mine
R => Rock

```python
sonar_data.groupby(60).mean()
```

```python
# seperating data and Labels
X = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]
```

```python
print(X)
print(y)
```

Training and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
training_data_accuracy
```

```python
X_test_pred = model.predict(X_test)
training_data_accuracy = accuracy_score(X_test_pred, Y_test)
training_data_accuracy
```
